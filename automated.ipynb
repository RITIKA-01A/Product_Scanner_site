````python
import cv2
import json
import threading
import google.generativeai as genai
from PIL import Image

# üîê Gemini Config
genai.configure(api_key="AIzaSyB9gKOT0r4k73X5i-Gt6JNpfLMxWq3HUa8")
model = genai.GenerativeModel("gemini-2.0-flash")

# üìã Gemini Response Handler
def extract_product_info_from_image(frame):
    pil_image = Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    try:
        response = model.generate_content([
            "Extract the product details from this image in JSON format. Only return JSON with keys: name, brand, product_type, size, expiry, price, sku.",
            pil_image
        ])
        raw_text = response.text.strip()
        if raw_text.startswith("```json"):
            raw_text = raw_text.replace("```json", "").replace("```", "").strip()

        data = json.loads(raw_text)

        if isinstance(data, list):
            data = sorted(data, key=lambda x: sum(bool(v) for v in x.values()), reverse=True)[0]

        print("\nüßæ Extracted Product Info (JSON):\n")
        print(json.dumps(data, indent=4))

    except Exception as e:
        print("‚ùå Error processing Gemini response:", e)


# üé• Webcam Feed
cap = cv2.VideoCapture(1)
if not cap.isOpened():
    print("‚ùå Failed to open camera.")
    exit()

print("üì∑ Camera started. Press SPACE to capture an image...")

try:
    while True:
        ret, frame = cap.read()
        if not ret:
            print("‚ùå Failed to capture frame.")
            break

        cv2.imshow("Live Product Scanner (Press SPACE to Capture, Q to Quit)", frame)

        key = cv2.waitKey(1) & 0xFF

        if key == ord(' '):  # Spacebar pressed
            print("üîç Capturing image and extracting in background...")
            # Clone the frame and run processing in a new thread
            thread_frame = frame.copy()
            threading.Thread(target=extract_product_info_from_image, args=(thread_frame,)).start()

        elif key == ord('q'):
            print("üëã Exiting scanner...")
            break

except KeyboardInterrupt:
    print("‚õî Scanner interrupted by user.")

cap.release()
cv2.destroyAllWindows()
````
